# Accéléromètre au poignet : inclinaison à partir de l'accélération 3D

### 1) Définition des mots clés

- **Accélération :** L'accélération est définie comme la variation de la vitesse d'un objet par rapport au temps. Elle décrit comment la vitesse d'un objet change au fil du temps, que ce soit en termes de grandeur (accélération ou décélération) ou de direction.
  
L'accélération se calcule à partir de la formule suivante :
a=ΔtΔv​ exprimée en mètres par seconde carrée (m/s2)

où :
a est l'accélération,
Δv est la variation de vitesse (vitesse finale moins vitesse initiale),
Δt est le temps pendant lequel cette variation a lieu.

Il existe plusieurs types d'accélérations en physique, notamment :
    Accélération linéaire 
    Accélération centripète 
    Acceleration gravitationelle 
    Acceleration angulaire

- **Accéléromètre :** Un accéléromètre est un capteur ou un dispositif qui permet de mesurer l'accélération d'un objet. Plus précisément, il détecte les changements de vitesse (l'accélération linéaire ou angulaire) dans une ou plusieurs directions. L'accéléromètre est capable de mesurer aussi bien les accélérations causées par le mouvement que celles dues à la gravité.Il peut mesurer les accélérations dans une, deux ou trois dimensions, selon le modèle.

- **Inclinaison (du poignet):** L'inclinaison, de manière générale, désigne l'action ou l'état de pencher ou de dévier par rapport à une position verticale, horizontale ou normale. Par exemple, on peut parler de l'inclinaison d'un plan, d'un objet, d'une planète ou même d'une partie du corps. Dans notre cas nous allons nous concentrer sur l'inclinaison du poignet.
L'inclinaison du poignet fait référence à la capacité de ce dernier à bouger ou à s'incliner selon deux axes principaux :
    **Inclinaison radiale** : Il s'agit du mouvement qui amène la main vers le côté du pouce. En termes anatomiques, c'est l'inclinaison du poignet vers le radius, l'os de l'avant-bras situé du côté du pouce.
    **Inclinaison ulnaire** : C'est le mouvement qui amène la main vers le côté de l'auriculaire (petit doigt). Anatomiquement, cela correspond à une inclinaison du poignet vers l'ulna, l'os situé du côté de l'auriculaire.
Ces mouvements d'inclinaison du poignet sont rendus possibles par les articulations et les muscles situés autour du poignet, permettant une certaine flexibilité et mobilité de la main dans différentes directions.

CMST, Comment fonctionne le poignet, https://chirurgie-main-epaule.eu/comment-fonctionne-le-poignet/

- **3D :** La 3D, ou espace tridimensionnel, est définie par trois axes perpendiculaires entre eux : l'axe X (horizontal), l'axe Y (vertical) et l'axe Z (profondeur). Dans un espace 2D, un point est localisé par deux coordonnées correspondant à sa position sur les axes X et Y. En 3D, un troisième axe, l'axe Z, est ajouté, permettant de décrire la profondeur. Ainsi, la position d'un point dans cet espace est déterminée par trois coordonnées : une valeur sur l'axe X, une sur l'axe Y et une sur l'axe Z. 

Sciences du Sport, Description anatomique du mouvement, https://www.sci-sport.com/theorie/chapitre-2-description-anatomique-du-mouvement.php#coord2d3d

- **AVC :** Un **accident vasculaire cérébral (AVC)** est une urgence médicale qui se produit lorsqu'il y a une interruption soudaine de l'irrigation sanguine au cerveau. Cela prive les cellules cérébrales d'oxygène, ce qui les fait se détériorer et entraîne divers symptômes neurologiques. Chaque année, environ 15 millions de personnes dans le monde souffrent d'un AVC, avec des conséquences graves : un tiers des personnes en décède, un tiers souffre de handicaps permanents, et un tiers se remet complètement.

2. **AVC hémorragique (15%)** : causé par la rupture d'un vaisseau sanguin dans le cerveau, entraînant une fuite de sang dans le tissu cérébral. Cela peut résulter de diverses causes, y compris des anomalies vasculaires ou l'hypertension.

Les symptômes d'un AVC varient en fonction de la région du cerveau touchée, ce qui peut inclure des troubles de la parole, des paralysies, ou des problèmes de coordination. L’**aire motrice primaire**, par exemple, est essentielle pour le contrôle des mouvements, et son atteinte peut affecter la motricité des membres.

Il est crucial de reconnaître rapidement les signes d'un AVC et d'agir rapidement pour minimiser les dommages cérébraux.rveau.

- **Membre supérieur :** Les membres supérieurs désignent, en anatomie, l'ensemble des structures du corps humain situées au-dessus du tronc et qui permettent la mobilité et la manipulation d'objets. Ils incluent les épaules, les bras, les avant-bras, les poignets et les mains.
Les fonctions principales des membres supérieurs sont la mobilité et la manipulation, le mouvement et la coordination.

### Contexte : Utilisation d’un accéléromètre au poignet pour le suivi de l’inclinaison post-AVC

Après un accident vasculaire cérébral (AVC), les patients souffrent souvent de troubles moteurs et de la coordination, affectant leur capacité à effectuer des mouvements précis et à maintenir certaines postures. Dans ce cadre, l’utilisation d’un accéléromètre placé au poignet se révèle être un outil efficace pour évaluer et surveiller la rééducation, en particulier pour analyser l’inclinaison du poignet et les mouvements du bras. 

Les accéléromètres mesurent l’accélération dans trois dimensions (x, y, z) et peuvent être utilisés pour détecter des inclinaisons anormales ou des limitations dans l’amplitude de mouvement. Cela permet de suivre les progrès des patients au fil du temps, de mieux comprendre leurs capacités fonctionnelles résiduelles, et de guider la rééducation.

### Inclinaison et rééducation post-AVC

Lors de la rééducation post-AVC, la capacité à maintenir une inclinaison correcte du poignet ou à effectuer des gestes précis est souvent altérée en raison d'une faiblesse musculaire ou d'une perte de coordination. En utilisant un accéléromètre au poignet, on peut suivre l’inclinaison du poignet par rapport à l’axe vertical, un indicateur utile pour évaluer la qualité de certains gestes essentiels dans la vie quotidienne, tels que la prise d’objets ou l'extension du bras.

### Limites et perspectives

Dans la réalité, le capteur au poignet peut être orienté de manière aléatoire en fonction des mouvements naturels du patient, rendant parfois difficile l’interprétation des données. De plus, si le patient effectue des mouvements rapides ou saccadés, l’accélération mesurée ne correspond plus seulement à l’accélération due à la gravité, mais inclut également l’accélération résultant du mouvement. Pour atténuer ces difficultés, les analyses sont souvent effectuées dans des conditions quasi stalimitantou en filtrant les mouvements brusques pour mieux isoler les variations dues à l’inclinaison du poignet.

En conclusion, l’utilisation d’un accéléromètre pour surveiller l’inclinaison du poignet chez les patients post-AVC offre une approche quantitative et objective pour évaluer les capacités motrices et ajuster les programmes de rééducation. L'analyse de ces données permet de détecter des tendances positives ou négatives dans la récupération fonctionnelle et de mieux comprendre la réadaptation motrice du patient.
