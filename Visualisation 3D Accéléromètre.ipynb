### Visualisation dans l'espace 3D

Pour représenter les axes X, Y, Z et la gravité à différentes inclinaisons du poignet, nous allons simuler plusieurs positions où le poignet est incliné par rapport à la gravité. À chaque inclinaison, les axes X, Y, et Z vont pivoter par rapport à la position neutre du poignet.

Voici plusieurs représentations du poignet à différentes inclinaisons, avec les axes X, Y, Z et la gravité :

Poignet en position neutre : Le poignet est à plat, les axes X, Y et Z sont alignés comme dans le schéma précédent.

Poignet incliné selon l'axe X : Le poignet est incliné vers le haut ou vers le bas, modifiant l'orientation de l'axe Z par rapport à la gravité.

Poignet incliné selon l'axe Y : Le poignet est incliné vers la gauche ou la droite, modifiant l'axe X.

Poignet tordu selon l'axe Z : Le poignet est tordu, affectant l'orientation des axes X et Y.
